```python
# DataFrame can be also summarised using Data Aggregation, which is the process of combining data from multiple sources
# and organising it into a view that can be used for further analysis 
# The custom aggregation function can be created, where any function can be added that aggregates an array to the agg method

# resource: https://www.cnblogs.com/johnyang/p/12793520.html

grouped=df.groupby('Gender')
def max_min(arr):
    return arr.max()-arr.min()
grouped.agg(max_min)
```

```python
# The summary of the numerical data for our group 'Gender'

grouped.describe()
```

```python
# Additionaly quantilies can be applied to the DataFrame
# For 'Gender' and 'Purchased' group we can apply 25% based on 'Annual Salary'

df['q'] = df.groupby(['Gender', 'Purchased'])['AnnualSalary'].transform(lambda x: x.quantile(0.25))
df
```

```python
# The DataFrame can also be aggregates with multiple fuunctions at once
# To show this the new DataFrame and new group based on 'Gender' and 'Purchased' can be created

# resource: https://www.cnblogs.com/johnyang/p/12793520.html

df['df_new']=df['Age']/df['AnnualSalary']
grouped=df.groupby(['Gender','Purchased'])
group_new=grouped['df_new']
grouped.size()
```

```python
# First 5 rows from the new data

df.head()
```

```python
# From new data we can select and show only top ten values for the new group

def top(df,n=10,column='df_new'):
    return df.sort_values(by=column)[-n:]
df.sort_values(by='df_new')[-10:]
```

```python
# The new aggregated group can be further analysed
# The average value from the new aggregated group

group_new.agg('mean')
```

```python
# Return from multiple functions applied to the new aggregated group

group_new.agg(['std','mean',max_min])
```

```python
# Names of functions in the new aggregated group can be changed to custom names for a nicer presentation

group_new.agg([('Average','mean'),('Standard',np.std)])
```

```python
# There is an option to specify either a list of functions which then applies to all columns or 
# Different functions per column
# Two columns from the new DataFrame can be summarised using the same four functions

functions=['count','mean','max','min']
result=grouped['df_new','AnnualSalary'].agg(functions)
result
```

```python
# Column names can also be combined together and presented as the key argument
# This can be done using a list of tuples and custom names for columns

list_tuples=[('Average','mean'),('Variance',np.var)]
grouped['df_new','AnnualSalary'].agg(list_tuples)
```

```python
# Different functions can be applied to different columns also
# Maximum 'Age' can be compared with minimum 'Annual Salary'

grouped.agg({'Age':np.max,'AnnualSalary':'min'})
```

```python
# The list of tuples summary with all standard functions

grouped.agg(list_tuples)
```

```python
# The same list of tuples can be compared using only chosen functions, which can differ by column

grouped.agg({'df_new':['min','max','mean','std'],'User ID':'count'})
```

```python
# We can also show top three column entries for a 'Gender' and 'Purchased' group

df.groupby(['Purchased','Gender']).apply(top,n=3,column='AnnualSalary')
```
